```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

df=pd.read_csv("churn-bigml-20.csv")
print(df.head())
```

```python
plt.figure(figsize=(20,10))
sb.countplot(x=df['State'], hue=df['Churn'])
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.title('Churn by State')
plt.show()
```

```python


plt.figure(figsize=(15, 5))
sb.scatterplot(x=df['Total day calls'], y=df['Total day charge'], hue=df['State'], data=df, s=50)
plt.xlabel('Total day calls')
plt.ylabel('Total day charge')
plt.title('Total day calls vs Total day charge by State')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
#plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,5))
sb.scatterplot(x=df['Total night calls'], y=df['Total night charge'], hue=df['State'], data=df, s=50)
plt.xlabel('Total night calls')
plt.ylabel('Total night charge')
plt.title('Total night calls vs Total night charge by State')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
#plt.tight_layout()
plt.show()
```
